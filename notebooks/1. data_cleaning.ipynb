## Transformations

I only need a column with the salary information and I would like it to be in Euros, which is more suitable for my target audience.

```python
df['salary_in_euros'] = df['salary_in_usd']*0.92
df.head()
```

```python
df['salary_in_euros'] = df['salary_in_euros'].astype(int)
df.head()
```
